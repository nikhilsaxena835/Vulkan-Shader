```python
from tqdm.notebook import tqdm
import time

with tqdm(total=1, desc="Installing dependencies") as pbar:
    !pip install --upgrade ultralytics onnx onnxsim
    pbar.update(1)

from ultralytics import YOLO

with tqdm(total=1, desc="Loading model") as pbar:
    model = YOLO('yolov8s-seg.pt')  # or yolov8n-seg.pt
    pbar.update(1)

with tqdm(total=100, desc="Exporting to ONNX") as pbar:
    onnx_model = model.export(format='onnx', opset=12, imgsz=640)
    for _ in range(99):
        time.sleep(0.01)
        pbar.update(1)

print("Export complete! File saved as:", onnx_model)
```
